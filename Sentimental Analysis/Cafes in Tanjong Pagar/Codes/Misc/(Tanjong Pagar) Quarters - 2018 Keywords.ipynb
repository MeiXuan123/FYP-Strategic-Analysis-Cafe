```python
import pandas as pd
import datetime
import math
import numpy

Burpple = pd.read_excel(r"(TP) From2016_Sentimental_Burpple.xlsx")
Dates = Dates = Burpple['Date'].tolist()

condition = ''
Form = []
why = []
day = ''
mon = ''
year = ''
monyear = ''
Date = ''

for i in Dates:
    
    condition = i.split()[2]
    
    if condition == "ago":
        day = i.split()[0]
        monyear = "May 2018"
        year = "2018"
        Date = day + " " + monyear
        Form.append(Date)
        why.append(year)
        
    elif condition == "at":
        mon = i.split()[0]
        day = i.split()[1]
        year = "2018"
        Date = day+" "+ mon + " 2018"
        Form.append(Date)
        why.append(year)
        
    else:
        day = i.split()[1]
        day = day.replace(',','')
        mon = i.split()[0]
        year = i.split()[2]
        Date = day + " " + mon + " " + year
        Form.append(Date)
        why.append(year)
        
 
              
num = len(Burpple)
NewDate = []
Quart = []
Q = 0
yy = ''
strdate = ''

for i in range (0, num):
    Date = datetime.datetime.strptime(Form[i], '%d %b %Y').strftime('%d/%m/%Y')
    Date = datetime.datetime.strptime(Date,'%d/%m/%Y')
    
    NewDate.append(Date)

    Q=math.ceil(Date.month/3.)
    #print(Q)
    Quart.append(Q)
    
    #strdate = str(Date)
    #yy = Date.Year
    #bpyear.append(yy)
    
       
FormatDate = pd.Series(NewDate)
Burpple['Formated_Date'] = FormatDate.values

Quarter = pd.Series(Quart)
Burpple['Quarter'] = Quarter.values

Year = pd.Series(why)
Burpple['Year'] = Year.values

ex = pd.ExcelWriter('(TP) Quarters.xlsx', options={'encoding':'utf-8'})
Burpple.to_excel(ex, 'Data', index=False)

#AllNames = Burpple.groupby('Caf√©_Names', sort=False, as_index=False).Title.count().reset_index(drop=True)
#NoR = AllNames['Title'].tolist()


### sorting the data by Years

y2016 = pd.DataFrame()
y2017 = pd.DataFrame()
y2018 = pd.DataFrame()
No2016 = []
No2017 = []
No2018 = []

for i in range (0, num):
    if why[i] == '2016':
        No2016.append(i)
    if why[i] == '2017':
        No2017.append(i)
    if why[i] == '2018':
        No2018.append(i)
    
y2016 = pd.DataFrame(Burpple.iloc[No2016])
y2017 = pd.DataFrame(Burpple.iloc[No2017])
y2018 = pd.DataFrame(Burpple.iloc[No2018])


y2016 = y2016.reset_index(drop = True)
y2017 = y2017.reset_index(drop = True)
y2018 = y2018.reset_index(drop = True)

#next decide how you want to split the data.
#seperate each year into diff files? How about the average sentiment?

###  Resturant  Avg_sentiment Quarter_1 Quarter_2 Quarter_3 Quarter_4
###  xxx        0.01          0.2       0.021     0.45      0.46
```

```python
# for Quarter 1

keywords = ['ice cream', 'onsen egg', 'french toast', 'scrambled egg', 'poached egg']
num = len(keywords)
rev = q1['Cleaned_Reviews'].tolist()
rev1 = q1['Review'].tolist()
polarity = q1['polarity'].tolist()
num1 = len(rev)
temprev = ""
nothing = []
count = 0
total = 0
Avg = 0
count_p = []
total_p = []
avg_p = []
maxneg = []
maxn = 1
maxp = -1
absn = 1
maxnrev = []
maxprev = []
absnrev = []
maxnpor = []
maxppor = []
absnpor = []
non = 0
noneu = 0
nopov = 0

for i in range(0, num):
    wek = str(keywords[i])
    for x in range(0, num1):
        temprev = str(rev[x])
        if temprev.find(wek) == -1:
            nothing.append
        else:
            total = total + polarity[x]
            count = count + 1
            if polarity[x] < maxn:
                maxn = polarity[x]
                non = x
            elif polarity[x] > maxp:
                maxp = polarity[x]
                noneu = x
            elif abs(polarity[x]) < absn:
                absn = polarity[x]
                nopov = x
    if total == 0 or count == 0:
        Avg = 0
    else:    
        Avg = total / count
    maxnrev.append(rev1[non])
    maxprev.append(rev1[noneu])
    absnrev.append(rev1[nopov])
    maxnpor.append(maxn)
    maxppor.append(maxp)
    absnpor.append(absn)
    avg_p.append(Avg)
    count_p.append(count)
    total_p.append(total)
    Avg = 0
    count = 0
    total = 0
    maxn = 1
    maxp = -1
    absn = 1
    non = 0
    noneu = 0
    nopov = 0
    
Final = pd.DataFrame() 

keywords = pd.Series(keywords)
Final['Food_Items'] = keywords.values

count_p = pd.Series(count_p)
Final['Count'] = count_p.values

maxnrev = pd.Series(maxnrev)
Final['Negative Reviews'] = maxnrev.values

absnrev = pd.Series(absnrev)
Final['Neutal Reviews'] = absnrev.values

maxprev = pd.Series(maxprev)
Final['Positive Reviews'] = maxprev.values

maxnpor
maxppor
absnpor

maxnpor = pd.Series(maxnpor)
Final['Polarity of Negative Reviews'] = maxnpor.values

absnpor = pd.Series(absnpor)
Final['Polarity of Neutral'] = absnpor.values

maxppor = pd.Series(maxppor)
Final['Polarity off Positive Reviews'] = maxppor.values


Final

#ex = pd.ExcelWriter('(TP) AS for FoodItems 2018Q1.xlsx', options={'encoding':'utf-8'})
#Final.to_excel(ex, 'Data', index=False)
```
