```python
import pandas as pd
import datetime
import math
import numpy

Burpple = pd.read_excel(r"(TP) From2016_Sentimental_Burpple.xlsx")
Dates = Dates = Burpple['Date'].tolist()

condition = ''
Form = []
why = []
day = ''
mon = ''
year = ''
monyear = ''
Date = ''

for i in Dates:
    
    condition = i.split()[2]
    
    if condition == "ago":
        day = i.split()[0]
        monyear = "May 2018"
        year = "2018"
        Date = day + " " + monyear
        Form.append(Date)
        why.append(year)
        
    elif condition == "at":
        mon = i.split()[0]
        day = i.split()[1]
        year = "2018"
        Date = day+" "+ mon + " 2018"
        Form.append(Date)
        why.append(year)
        
    else:
        day = i.split()[1]
        day = day.replace(',','')
        mon = i.split()[0]
        year = i.split()[2]
        Date = day + " " + mon + " " + year
        Form.append(Date)
        why.append(year)
        
 
              
num = len(Burpple)
NewDate = []
Quart = []
Q = 0
yy = ''
strdate = ''

for i in range (0, num):
    Date = datetime.datetime.strptime(Form[i], '%d %b %Y').strftime('%d/%m/%Y')
    Date = datetime.datetime.strptime(Date,'%d/%m/%Y')
    
    NewDate.append(Date)

    Q=math.ceil(Date.month/3.)
    #print(Q)
    Quart.append(Q)
    
    #strdate = str(Date)
    #yy = Date.Year
    #bpyear.append(yy)
    
       
FormatDate = pd.Series(NewDate)
Burpple['Formated_Date'] = FormatDate.values

Quarter = pd.Series(Quart)
Burpple['Quarter'] = Quarter.values

Year = pd.Series(why)
Burpple['Year'] = Year.values

ex = pd.ExcelWriter('(TP) Quarters.xlsx', options={'encoding':'utf-8'})
Burpple.to_excel(ex, 'Data', index=False)

#AllNames = Burpple.groupby('Café_Names', sort=False, as_index=False).Title.count().reset_index(drop=True)
#NoR = AllNames['Title'].tolist()


### sorting the data by Years

y2016 = pd.DataFrame()
y2017 = pd.DataFrame()
y2018 = pd.DataFrame()
No2016 = []
No2017 = []
No2018 = []

for i in range (0, num):
    if why[i] == '2016':
        No2016.append(i)
    if why[i] == '2017':
        No2017.append(i)
    if why[i] == '2018':
        No2018.append(i)
    
y2016 = pd.DataFrame(Burpple.iloc[No2016])
y2017 = pd.DataFrame(Burpple.iloc[No2017])
y2018 = pd.DataFrame(Burpple.iloc[No2018])


y2016 = y2016.reset_index(drop = True)
y2017 = y2017.reset_index(drop = True)
y2018 = y2018.reset_index(drop = True)

#next decide how you want to split the data.
#seperate each year into diff files? How about the average sentiment?

###  Resturant  Avg_sentiment Quarter_1 Quarter_2 Quarter_3 Quarter_4
###  xxx        0.01          0.2       0.021     0.45      0.46
```

```python
### for the year 2018

AllNames2018 = y2018.groupby('Café_Names', sort=False, as_index=False).Title.count().reset_index(drop=True)
NoR2018 = AllNames2018['Title'].tolist()

Quarter = y2018['Quarter'].tolist()

num = len(y2018)
No1 = []
No2 =[]
No3 =[]
No4 =[]

for i in range(0, num):
    if Quarter[i] == 1:
        No1.append(i)
    if Quarter[i] == 2:
        No2.append(i)
    if Quarter[i] == 3:
        No3.append(i)
    if Quarter[i] == 4:
        No4.append(i)
        
#creating databases based on the quarter

q1 = pd.DataFrame(y2018.iloc[No1])
q2 = pd.DataFrame(y2018.iloc[No2])
q3 = pd.DataFrame(y2018.iloc[No3])
q4 = pd.DataFrame(y2018.iloc[No4])

# Quarter 1

AllNames1 = q1.groupby('Café_Names', sort=False, as_index=False).Title.count().reset_index(drop=True)
NoR1 = AllNames1['Title'].tolist()

Q1YN =[]
names = AllNames2018['Café_Names'].tolist()
names1 = AllNames1['Café_Names'].tolist()

for i in names:
    if i in names1:
        Q1YN.append("Yes")
    else:
        Q1YN.append("")

polarity = q1['polarity'].tolist()        
num = len(NoR2018)
pos = 0
poss = 0
counta = 0
total = float(0)
average = float(0)
avg1 = []

for i in range (0, num):
    if Q1YN[i] == "Yes":
        pos = pos + NoR1[counta]
        total = 0
        average = 0
        for x in range(poss, pos):
            total = total + polarity[x]
        average = total / NoR1[counta]
        avg1.append(average)
        poss = pos
        counta = counta + 1
    else:
        avg1.append(0)
        
Avg_Q1 = pd.Series(avg1)       
AllNames2018['Q1_Average_Polarity'] = Avg_Q1.values
```
