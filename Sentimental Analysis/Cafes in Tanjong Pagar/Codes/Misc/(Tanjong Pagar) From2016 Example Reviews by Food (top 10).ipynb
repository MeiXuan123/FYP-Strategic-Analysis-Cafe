```python
import pandas as pd
import datetime
import math
import numpy

Burpple = pd.read_excel(r"(TP) From2016_Sentimental_Burpple.xlsx")
Dates = Dates = Burpple['Date'].tolist()

```

```python
keywords = ['ice cream', 'onsen egg', 'french toast', 'scrambled egg',
           'poached egg', 'earl grey','poached eggs',
           'beef cheek', 'ya tou', 'braised beef']
num = len(keywords)
rev = Burpple['Cleaned_Reviews'].tolist()
rev1 = Burpple['Review'].tolist()
polarity = Burpple['polarity'].tolist()
num1 = len(rev)
temprev = ""
nothing = []
count = 0
total = 0
Avg = 0
count_p = []
total_p = []
avg_p = []
maxneg = []
maxn = 1
maxp = -1
absn = 1
maxnrev = []
maxprev = []
absnrev = []
maxnpor = []
maxppor = []
absnpor = []
non = 0
noneu = 0
nopov = 0

for i in range(0, num):
    wek = str(keywords[i])
    for x in range(0, num1):
        temprev = str(rev[x])
        if temprev.find(wek) == -1:
            nothing.append
        else:
            total = total + polarity[x]
            count = count + 1
            if polarity[x] < maxn:
                maxn = polarity[x]
                non = x
            elif polarity[x] > maxp:
                maxp = polarity[x]
                noneu = x
            elif abs(polarity[x]) < absn:
                absn = polarity[x]
                nopov = x
    if total == 0 or count == 0:
        Avg = 0
    else:    
        Avg = total / count
    maxnrev.append(rev1[non])
    maxprev.append(rev1[noneu])
    absnrev.append(rev1[nopov])
    maxnpor.append(maxn)
    maxppor.append(maxp)
    absnpor.append(absn)
    avg_p.append(Avg)
    count_p.append(count)
    total_p.append(total)
    Avg = 0
    count = 0
    total = 0
    maxn = 1
    maxp = -1
    absn = 1
    non = 0
    noneu = 0
    nopov = 0
    
Final = pd.DataFrame() 

keywords = pd.Series(keywords)
Final['Food_Items'] = keywords.values

count_p = pd.Series(count_p)
Final['Count'] = count_p.values

maxnrev = pd.Series(maxnrev)
Final['Negative Reviews'] = maxnrev.values

absnrev = pd.Series(absnrev)
Final['Neutal Reviews'] = absnrev.values

maxprev = pd.Series(maxprev)
Final['Positive Reviews'] = maxprev.values

maxnpor
maxppor
absnpor

maxnpor = pd.Series(maxnpor)
Final['Polarity of Negative Reviews'] = maxnpor.values

absnpor = pd.Series(absnpor)
Final['Polarity of Neutral'] = absnpor.values

maxppor = pd.Series(maxppor)
Final['Polarity off Positive Reviews'] = maxppor.values


Final

ex = pd.ExcelWriter('(TP) From2016 Example Reviews by Food (Top 10).xlsx', options={'encoding':'utf-8'})
Final.to_excel(ex, 'Data', index=False)
```
