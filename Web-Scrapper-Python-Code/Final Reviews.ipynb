```python
import time
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import requests as rq
import re
from time import sleep
from random import randint
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, WebDriverException, NoSuchElementException
from IPython.core.display import clear_output


title = []
body = []
revcafe = []
date = []

xl_workbook = pd.ExcelFile('C:/Users/Mei Xuan/Downloads/CompiledCafeLinks.xlsx')  # Load the excel workbook
##change accordingly to where you downloaded the excel sheet

df = xl_workbook.parse("Data")  # Parse the sheet into a dataframe #'Data' because i name my Sheet1 as 'Data'
urllist = df['Links'].tolist()

splitlist = []
for sl in urllist:
    splitlist.append(sl.split('?')[0])


for c_url in splitlist:
    
    #Print out which url python is currently processing
    print("Processing Individual Cafes' URL: {}".format('https://www.burpple.com' + c_url + '/reviews' ))
    clear_output(wait = True)
    
    #pause the loop awhile, 1 to 4seconds
    sleep(randint(2,5))
    
    driver = webdriver.Chrome()
    driver.get('https://www.burpple.com' + c_url  + '/reviews')


###################################

##Review Page

    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    match=False
    while(match==False):
                    sleep(randint(6,9))
                    lastCount = lenOfPage
                    
                    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
                    sleep(randint(7,10))
                    if lastCount==lenOfPage:
                        match=True

#Now that the page is fully scrolled, grab the source code.
    source_data = driver.page_source

#Throw the source into BeautifulSoup and start parsing
    soup2 = bs(source_data,'lxml')

    revborder = soup2.find_all("div",{"class": "food card feed-item"})

    for rev in revborder:
        if (rev.find("div",{"class": "food-description-body"})) is not None:  ##cannot dont have a review body 
            
            try:
                date_cont = rev.find("div",{"class":"card-item-set--link-details"})
                date_elem = date_cont.find("div",{"class": "card-item-set--link-subtitle"}).text
                date_strip = date_elem.strip('\n\r\t": ')
            
            
            except AttributeError:
                date_strip = '-'
            
            try:
                title_cont = rev.find("div",{"class": "food-content card-item"})
                title_elem = title_cont.find("div",{"class": "food-description-title"}).text
            
            except AttributeError:
                title_elem = '-'
            
            try:
                body_cont = rev.find("div",{"class": "food-content card-item"})
                body_elem = body_cont.find("div",{"class": "food-description-body"}).text
                body_strip = body_elem.strip('\n\r\t":  ')
                
            except AttributeError:
                body_strip = '-'
            
            
            title.append(title_elem)
            body.append(body_strip)
            date.append(date_strip)
    
    ##must not be inside the for loop!!!!!!!!!!!!!
    ##if not later you see your cafenames come out in thousands
    
#name of cafe
    revborder2 = soup2.find_all("div",{"class": "food-description-body"})
    rc = soup2.find("h1").text.strip('\n\r\t": ')    
    y = len(revborder2)
    for x in range(0,y):
        n = rc
        revcafe.append(n)
    
    
##Print after everything is done
    print('Done Scraping for this page')
        
```

```python
BurpReviews = pd.DataFrame({
        "Cafe_Names" : revcafe,
        "Title" : title,
        "Review" : body,
        "Date" : date})

BurpReviews = BurpReviews[['Cafe_Names','Date', 'Title','Review']]

BurpReviews
```

```python
ex = pd.ExcelWriter('BurpReviews.xlsx', options={'encoding':'utf-8'})
BurpReviews.to_excel(ex, 'Data', index=False)
```
