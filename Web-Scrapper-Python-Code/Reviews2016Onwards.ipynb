```python
import xlrd
import xlwt
import openpyxl  

sheet_data = []   

wb = xlrd.open_workbook('C:/Users/Mei Xuan/Downloads/OverallReviews.xlsx')  #where you download your excel files at
p = wb.sheet_names()
for y in p:
    sh = wb.sheet_by_name(u'Data')
    nrows = sh.nrows
    for rownum in range(1,nrows): #i dont want the first line which is the header
        sheet_data.append((sh.row_values(rownum)))
        
bad_words = ['2012','2013','2014','2015']  #change according to pref
cafename = []
date = []
title = []
rev = []
rows_deleted = []

for i in sheet_data:
    if not any(bad_word in i[1] for bad_word in bad_words):
        cafename.append(i[0])
        date.append(i[1])
        title.append(i[2])
        rev.append(i[3])
    else:
        rows_deleted.append(i)
        

```

```python
import pandas as pd

BurpReviewDate = pd.DataFrame({
        "Cafe_Names" : cafename,
        "Title" : title,
        "Review" : rev,
        "Date" : date})

BurpReviewDate = BurpReviewDate[['Cafe_Names','Date', 'Title','Review']]

BurpReviewDate
```

```python
ex = pd.ExcelWriter('ReviewsDate2016Onwards.xlsx', options={'encoding':'utf-8'})
BurpReviewDate.to_excel(ex, 'Data', index=False)
```
